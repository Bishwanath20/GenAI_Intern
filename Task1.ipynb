```python
"""Problem Statement:
Given a username and password, check whether login is successful.

Requirements:

Print "Login Successful" if both username and password match

Otherwise print "Invalid Credentials"

Real-World Application: Authentication systems
"""
```

```python
# 1. User Logic Check

given_username = "admin"
given_password = "1234"

username = "admin"
password = "1234"

if username == given_username and password == given_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
"""
Problem Statement:
Analyze student results.

Requirements:

A student passes if marks ≥ 50

Count total pass students

Count total fail students

Print final result clearly

Real-World Application: Academic evaluation systems
"""
```

```python
# 2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Number of Pass Students:", pass_count)
print("Number of Fail Students:", fail_count)
```

```python
"""
Problem Statement:
Clean and standardize user names.

Requirements:

Remove extra spaces

Convert all names to lowercase

Print cleaned list

Real-World Application: Data preprocessing before analysis
"""
```

```python
#3 Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
"""
Problem Statement:
Analyze message sizes.

Requirements:

Print length of each message

Flag messages longer than 10 characters

Real-World Application: Text filtering and validation systems
"""
```

```python
#4 Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}', Length: {length}")

    if length > 10:
        print("This message is longer than 10 characters")
```

```python
"""
Problem Statement:
Detect error messages from system logs.

Requirements:

Count number of "ERROR" entries

Real-World Application: Log monitoring systems
"""
```

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")
print("Total ERROR entries:", error_count)
```
