```python
"""
Problem Statement:
Given a sentence, identify all unique words.

Requirements:

Count unique words

Display unique words

Real-Time Applications:
Text analytics, NLP preprocessing, search indexing, chat analysis
"""
```

```python
#1. Unique Words in a Sentence

sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
"""
Problem Statement:
Determine which employee has the highest salary.

Requirements:

Store data in dictionary

Identify highest salary

Real-Time Applications:
Payroll systems, HR analytics
"""
```

```python
#2. Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
"""
Problem Statement:
Find maximum and minimum from a list.

Real-Time Applications:
Sales analysis, monitoring systems
"""
```

```python
#3. Find Maximum and Minimum

numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
"""
Problem Statement:
Count products with price greater than 1000.

Real-Time Applications:
E-commerce filtering, budget analysis
"""
```

```python
#4. Count Products Above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
"""
Problem Statement:
Calculate attendance percentage.

Real-Time Applications:
HR systems, internship tracking
"""
```

```python
#5. Attendance Percentage Calculation

attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
"""
Problem Statement:
Count frequency of each character in a string.

Real-Time Applications:
Text analysis, password validation
"""
```

```python
#7. Count Character Frequency

text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character frequency:", frequency)
```

```python
"""
Problem Statement:
Calculate average marks from list.

Real-Time Applications:
Student result processing
"""
```

```python
#10. Calculate Average Marks

marks = [70, 80, 65, 75]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print("Average Marks:", average)
```
