```python
"""
Problem Statement:
Identify employees eligible for top performance bonus.

Requirements:

Find highest performance score

Handle tie cases

Display all top performers

Real-Time Application:
HR performance evaluation systems
"""
```

```python
#1. Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
"""
Problem Statement:
Analyze search query keywords.

Requirements:

Convert to lowercase

Remove punctuation

Count word frequency

Display words appearing more than once

Real-Time Application:
Search engines, NLP preprocessing
```

```python
#2. Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"
query = query.lower()

import string
rule = str.maketrans('', '', string.punctuation)
clean_query = query.translate(rule)

words = clean_query.split()
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1
result = {word: count for word, count in word_count.items() if count > 1}
print(result)
```

```python
"""
Problem Statement:
Validate sensor readings.

Requirements:

Identify even readings

Store (hour_index, value) pairs

Ignore odd readings

Real-Time Application:
Industrial monitoring systems
"""
```

```python
#3. Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
"""
Problem Statement:
Analyze email domain usage.

Requirements:

Count users per domain

Calculate percentage usage

Real-Time Application:
User analytics systems
"""
```

```python
#4. Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domains = [email.split('@')[1] for email in emails]
for domain in set(domains):
    count = domains.count(domain)
    percentage = (count / len(emails)) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
"""
Problem Statement:
Detect unusual sales spikes.

Requirements:

Calculate average sales

Detect sales 30% above average

Display day number and value

Real-Time Application:
Retail sales monitoring
```

```python
#5. Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales) / len(sales)
for day, sale in enumerate(sales, start=1):
    if sale > avg_sales * 1.30:
        print(f"Day {day}: {sale}")
```

```python
"""Problem Statement:
Identify duplicate user IDs.

Requirements:

Detect duplicates

Display frequency of duplicates

Real-Time Application:
Data integrity management systems
"""
```

```python
#6. Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
unique_ids = set(user_ids)

for uid in unique_ids:
    count = user_ids.count(uid)
    if count > 1:
        print(f"{uid} → {count} times")
```
