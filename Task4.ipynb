```python
"""
Problem Statement:
Validate a mobile recharge request.

Rules:

Recharge amount ≥ ₹50

Valid plans: 199, 299, 399, 599

Allow retry if invalid

Real-Time Use:
Telecom recharge apps, payment gateways
```

```python
#1. Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
            continue
        if amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid Plan. Available plans:", valid_plans)
            print("Please try again.\n")
validate_recharge()
```

```python
"""
Problem Statement:
Check stock levels and generate reorder alerts.

Rule:

If stock < 15 → Reorder Alert

Otherwise → Stock OK

Real-Time Use:
Warehouse systems, retail management
```

```python
#2. Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock OK (Stock = {stock})")
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 8,
    "Milk": 20
}

check_inventory(inventory)
```

```python
"""
Problem Statement:
Process student marks and determine pass/fail.

Rule:

Average ≥ 50 → Pass

Otherwise → Fail

Real-Time Use:
Online exam systems, evaluation portals
```

```python
#3. Student Result Processing System

def process_results(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"
student_marks = [60, 70, 45, 80]
result = process_results(student_marks)
print("Result:", result)
```

```python
"""
Problem Statement:
Calculate cab fare with peak hour charges.

Fare Rules:

Base fare = ₹50

₹12 per km

Peak hour → 25% extra

Allow retry

Real-Time Use:
Ride-hailing apps, transport systems
```

```python
#4. Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)
    if peak.lower() == "yes":
        fare += 0.25 * fare

    return fare


while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print(f"Total Fare: ₹{total_fare}")

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
"""
Problem Statement:
Check eligibility based on attendance percentage.

Rule:

≥ 75% → Eligible

Otherwise → Not Eligible

Real-Time Use:
HR payroll systems, internship tracking
```

```python
#5. Attendance Eligibility Checker

def check_eligibility(attendance_list):
    present_count = 0
    for day in attendance_list:
        if day == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_eligibility(attendance)
print("Status:", status)
```

```python
"""
Problem Statement:
Validate password strength.

Rules:

Minimum length = 8

At least one digit

At least one special character (@#$)

Real-Time Use:
Login systems, security validation
```

```python
#6. Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"
user_password = input("Enter password: ")
print(check_password(user_password))
```
