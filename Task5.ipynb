```python
"""
Problem Statement:
Analyze total post engagement.

Rule:

Total likes ≥ 1000 → Viral Post

Otherwise → Normal Engagement

Real-Time Use:
Social media analytics systems
```

```python
#1. Social Media – Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)
likes = [200, 300, 150, 400]
analyze_engagement(likes)
```

```python
"""
Problem Statement:
Check whether rainfall is adequate.

Rule:

Average rainfall ≥ required level → Adequate

Otherwise → Inadequate

Real-Time Use:
Agricultural monitoring systems
```

```python
#3. Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_data, required_level):

    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)
rainfall = [70, 75, 68, 80]
rainfall_checker(rainfall, 72)
```

```python
"""
Problem Statement:
Detect duplicate user accounts.

Rule:

If duplicates exist → Yes

Otherwise → No

Real-Time Use:
User account management systems
```

```python
#4. Social Media – Duplicate Account Detection

def detect_duplicates(usernames):

    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

users = ["bishwanath", "John", "bishwanath", "mikitha"]
detect_duplicates(users)
```

```python
"""
Problem Statement:
Check exam eligibility based on attendance.

Rule:

Attendance ≥ 75% → Eligible

Otherwise → Not Eligible

Real-Time Use:
Online education systems
```

```python
#9. E-Learning – Student Attendance Eligibility System

def attendance_checker(attendance_list):

    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)
attendance = [1, 1, 0, 1, 1]
attendance_checker(attendance)
```

```python
"""
Problem Statement:
Calculate electricity bill based on usage slabs.

Billing Slabs:

First 100 units → ₹3/unit

Next 100 units → ₹5/unit

Above 200 units → ₹7/unit

Real-Time Use:
Smart billing systems
```

```python
#10. Smart Electricity Bill Analyzer

def electricity_bill(units):
    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Status:", usage_status)

electricity_bill(250)
```
