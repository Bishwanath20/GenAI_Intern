```python
"""
Problem 1: Smart Light Controller

Create a class SmartLight.

Initialize:

light_name
Default status = "OFF"

Create methods:
turn_on() → sets status to "ON"
turn_off() → sets status to "OFF"
display_status() → prints light name and status

Create an object and test it.
```

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")


light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

```python
"""
Problem 2: Employee ID Card System

Create a class Employee.

Initialize:
name
emp_id
department

Create method:
display_id_card() → prints employee details

Create an object and test it.
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


employee = Employee("Rahul", "EMP102", "IT")
employee.display_id_card()
```

```python
"""
Problem 3: Mobile Contact Record

Create a class Contact.
Initialize:
name
phone
Create method:
display_contact() → prints contact information

Create an object and test it.
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


contact = Contact("Anita", "9876543210")
contact.display_contact()
```

```python
"""
Problem 4: Product Price Tag Generator
Create a class Product.

Initialize:
product_name
price
Create method:
print_price_tag() → prints product name and price

Create an object and test it.
```

```python
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")


product = Product("Headphones", 2499)
product.print_price_tag()
```
